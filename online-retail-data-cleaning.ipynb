```python
# TODO 2.1: Check for missing values
print("Missing values per column:")
print(df.isna().sum().sort_values(ascending=False))
print("\nTotal missing cells:", df.isna().sum().sum())

# TODO 2.2: Check for duplicates
dup_count = df.duplicated().sum()
print("\nDuplicate rows:", dup_count)
if dup_count > 0:
    print("Example duplicate rows:")
    display(df[df.duplicated()].head())

# TODO 2.3: Check for invalid values


# TODO 2.3.1. Negative quantities
neg_qty = df[df['Quantity'] < 0]
print("\nNegative quantity rows:", len(neg_qty))
display(neg_qty.head())

#TODO 2.3.2 Negative prices
neg_price = df[df['UnitPrice'] < 0]
print("\nNegative UnitPrice rows:", len(neg_price))
display(neg_price.head())

# TODO 2.3.3 Zero prices
zero_price = df[df['UnitPrice'] == 0]
print("\nZero UnitPrice rows:", len(zero_price))
display(zero_price.head())

# TODO 2.4: Analyze cancelled orders
# Hint: Invoices starting with 'C' indicate cancellations
# Filter the invoice no that started with 'C'
cancelled = df[df['InvoiceNo'].astype(str).str.startswith('C', na=False)]
print("\nCancelled orders (InvoiceNo starts with 'C'):", len(cancelled))
display(cancelled.head())

print("\nCancelled orders - negative quantity rate:",
      (cancelled['Quantity'] < 0).mean())
print("Cancelled orders - avg quantity:",
      cancelled['Quantity'].mean())
```


## **Question**
1. Why might there be negative quantities? Should we remove them or handle them differently?


- Negative quantities typically represent returned or cancelled items rather than simple data errors. In retail transaction systems, returns are often recorded as negative quantities to offset previous sales. In this dataset, many of these negative values are associated with invoice numbers starting with "C", whichindicates cancellations. Because of this, negative quantities should not automatically be removed. Instead, they can be handled differently depending on the goal of the analysis, for example, kept and flagged when analyzing returns, or excluded when focusing only on completed sales.

## Data Cleaning
- Removing duplicates
- Handling missing data
- Handling Invalid data

**Note:** Complete each task below and use *print* statement to show your output

```python
# Keep a backup of the original dataframe
df_clean = df.copy()

# TODO 3.1: Remove Duplicate rows
before = len(df_clean)
df_clean = df_clean.drop_duplicates()
after = len(df_clean)

print("3.1 Remove duplicates")
print("Rows before:", before)
print("Rows after:", after)
print("Duplicates removed:", before - after)

# TODO 3.2: Handle missing CustomerID
missing_cust = df_clean['CustomerID'].isna().sum()

print("\n3.2 Missing CustomerID")
print("Missing CustomerID rows:", missing_cust)
print("Percent missing:", round(100 * missing_cust / len(df_clean), 2), "%")

# TODO 3.2.1: create a column name HasCustomerID for missing customers
df_clean['HasCustomerID'] = df_clean['CustomerID'].notna()

print("\n3.2.1 HasCustomerID column created")
print(df_clean['HasCustomerID'].value_counts())

# TODO 3.3: Separate cancelled orders
# Hint: Create a column IsCancelled cancellation flag
df_clean['IsCancelled'] = df_clean['InvoiceNo'].astype(str).str.startswith('C', na=False)

print("\n3.3 IsCancelled column created")
print(df_clean['IsCancelled'].value_counts())

# TODO 3.4: Filter the non canceled orders
df_non_cancelled = df_clean[~df_clean['IsCancelled']].copy()

print("\n3.4 Non-cancelled orders")
print("Rows before:", len(df_clean))
print("Rows after filtering cancellations:", len(df_non_cancelled))

# TODO 3.5: Remove invalid quantities and prices
invalid_qty = df_non_cancelled['Quantity'] <= 0
invalid_price = df_non_cancelled['UnitPrice'] <= 0

print("3.5 Invalid quantities (Quantity <= 0):", invalid_qty.sum())
print("3.5 Invalid prices (UnitPrice <= 0):", invalid_price.sum())
print("3.5 Total invalid rows:",
      (invalid_qty | invalid_price).sum())

print("\nExamples of invalid rows:")
display(df_non_cancelled[invalid_qty | invalid_price].head())

# TODO 3.6:Keep only positive quantities and prices
df_non_cancelled_valid = df_non_cancelled[
    (df_non_cancelled['Quantity'] > 0) &
    (df_non_cancelled['UnitPrice'] > 0)
].copy()

print("3.6 Rows before filtering:", len(df_non_cancelled))
print("3.6 Rows after keeping positive Quantity and UnitPrice:",
      len(df_non_cancelled_valid))

# TODO 3.7: Standardize text columns
# Hint: Strip whitespace and standardize case for Description &Country names
df_non_cancelled_valid['Description'] = (
    df_non_cancelled_valid['Description']
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

df_non_cancelled_valid['Country'] = (
    df_non_cancelled_valid['Country']
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

print("\n3.7 Standardized Description and Country")
display(df_non_cancelled_valid[['Description', 'Country']].head())
```


## **Question**
- What's the trade-off between removing rows with missing CustomerID vs. keeping them?


- Removing rows with missing CustomerID makes it easier to perform customer-based analysis, such as grouping purchases by customer or identifiying repeat buyers. However, doing so also removes legitimate transactions from the dataset, which can effect overallsales analysis. Keeping rows with missing CustomerID preserves all transaction data, but limits analyses that rely on linking purchases to specific customers. The choice depends on whether the focus is on customer behavior or over sales trends.

## Detect Inconsistencies

- **Note:** When we group data by ID, ALL records with same ID should have same attributes. If they don't there exist data quality problem!

```python
# TODO 4.1: Check if each StockCode has consistent product name
## For each StockCode, how many different product names exist?
## Print StockCodes with multiple names
name_counts = (
    df_non_cancelled_valid
    .groupby('StockCode')['Description']
    .nunique()
    .sort_values(ascending=False)
)

inconsistent_names = name_counts[name_counts > 1]

print("StockCodes with multiple product names:", len(inconsistent_names))
display(inconsistent_names.head(10))

# TODO 4.2: Check if each CustomerID has consistent Country
# For each customer, how many different countries?
country_counts = (
    df_non_cancelled_valid
    .dropna(subset=['CustomerID'])
    .groupby('CustomerID')['Country']
    .nunique()
    .sort_values(ascending=False)
)

inconsistent_countries = country_counts[country_counts > 1]

print("Customers with multiple countries:", len(inconsistent_countries))
display(inconsistent_countries.head(10))

# TODO 4.3: Check if each StockCode has consistent price
# For each product, what's the price variation?
# Hint: find mean, max, min  IQR
def iqr(series):
    return series.quantile(0.75) - series.quantile(0.25)

price_stats = (
    df_non_cancelled_valid
    .groupby('StockCode')['UnitPrice']
    .agg(
        count='count',
        mean='mean',
        min='min',
        max='max',
        iqr=iqr
    )
)

price_stats['range'] = price_stats['max'] - price_stats['min']

print("Price variation statistics:")
display(price_stats.head())

# TODO4.4: Display top 10 products with higher price variation
top10_price_variation = price_stats.sort_values(
    ['iqr', 'range'],
    ascending=False
).head(10)

print("Top 10 StockCodes with highest price variation:")
display(top10_price_variation)
```
